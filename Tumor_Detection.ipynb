```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
df.head()
```

```python
x = df.drop('diagnosis', axis=1)
x.head()
```

```python
y = df['diagnosis']
y
```

### Training the Dataset

```python
# Dividing the dataset into trained and test set 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
print("The Total data shape:", df.shape)
print("The Trained x_data shape:", x_train.shape)
print("The Test x_data shape:", x_test.shape)
print("The Trained y_data shape:", y_train.shape)
print("The Test y_data shape:", y_test.shape)
```

```python
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test =ss.transform(x_test)
```

### Applying the Random Forest Classifier 

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)

y_pred = rfc.predict(x_test)
print(f"Accuracy Score:{accuracy_score(y_test, y_pred)*100:.2f}%")
```
