1. Create a simple line plot 
- a. Plot the data.  
- b. Customize the plot by adding a title, axis labels, and a grid.


```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

# a. Plot the data
plt.plot(x, y, marker='o', linestyle='-', color='b')

# b. Customize the plot
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

# Show the plot
plt.show()
```

2. Create a bar graph to represent the marks scored by students in a subject:  
  a. Plot the data as a bar graph.  
  b. Customize the colors and add a title.


```python
import matplotlib.pyplot as plt

students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

# a. Plot the data as a bar graph
plt.bar(students, marks, color=['skyblue', 'lightgreen', 'salmon', 'orange'])

# b. Customize the plot
plt.title("Marks Scored by Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.ylim(0, 100)  # Optional: set y-axis limit
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

3. Create a pie chart to represent the percentage distribution of a companys revenue from different regions:

  
- a. Create a pie chart with the region names as labels.  
- b. Highlight the region with the highest revenue.

```python
import matplotlib.pyplot as plt

regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# a. Create a pie chart with labels
# b. Highlight the region with the highest revenue
explode = [0.1 if r == max(revenue) else 0 for r in revenue]  # explode highest revenue [0.1,0,0,0]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=90, explode=explode)

# Add a title
plt.title("Revenue Distribution by Region")

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular.
plt.show()
```

4. Generate a histogram to show the frequency distribution of randomly generated integers between 1 and 100 (sample size = 1000).

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random integers between 1 and 100
data = np.random.randint(1, 101, size=1000)

# Create the histogram
plt.hist(data, color='skyblue', edgecolor='black')

# Customize the plot
plt.title("Histogram of Random Integers (1100)")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
